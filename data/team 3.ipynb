```python
# Scatter Plot of BMI vs. Age
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Enable inline plotting (for Jupyter Notebook)
%matplotlib inline

# Define file path
file_path = r"C:\Users\Gulru\Team3_Estimation_of_Obesity_Levels\data\Preprocessed_ObesityDataSet_raw_and_data_sinthetic.csv"

# Load dataset
data = pd.read_csv(file_path)

# Check if 'Age', 'Height', and 'Weight' exist
if 'Age' in data.columns and 'Height' in data.columns and 'Weight' in data.columns:
    # Calculate BMI
    data['BMI'] = data['Weight'] / (data['Height'] ** 2)

    # Drop missing values
    data = data.dropna(subset=['Age', 'BMI'])

    # Create scatter plot
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data['Age'], y=data['BMI'], color='blue', alpha=0.6)

    # Add labels
    plt.title("Scatter Plot of BMI vs. Age")
    plt.xlabel("Age")
    plt.ylabel("BMI")

    # Show the plot
    plt.show()
else:
    print("Required columns 'Age', 'Height', or 'Weight' not found. Available columns:", data.columns.tolist())
```

```python
#Bar Chart for BMI by Gender
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define file path
file_path = r"C:\Users\Gulru\Team3_Estimation_of_Obesity_Levels\data\Preprocessed_ObesityDataSet_raw_and_data_sinthetic.csv"

# Load dataset
data = pd.read_csv(file_path)

# Check if required columns exist
if 'Gender' in data.columns and 'Height' in data.columns and 'Weight' in data.columns:
    # Calculate BMI
    data['BMI'] = data['Weight'] / (data['Height'] ** 2)

    # Create bar plot
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Gender', y='BMI', data=data, estimator=lambda x: sum(x) / len(x), ci=None)

    # Add labels
    plt.title('Average BMI by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Average BMI')

    # Show plot
    plt.show()
else:
    print("Required columns 'Gender', 'Height', or 'Weight' not found. Available columns:", data.columns.tolist())
```

```python
#Boxplot of BMI by Frequency of Physical Activity
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define file path
file_path = r"C:\Users\Gulru\Team3_Estimation_of_Obesity_Levels\data\Preprocessed_ObesityDataSet_raw_and_data_sinthetic.csv"

# Load dataset
data = pd.read_csv(file_path)

# Check available columns
print("Available columns:", data.columns.tolist())

# Identify the correct column for physical activity
physical_activity_col = None
for col in data.columns:
    if "FAF" in col:  # 'FAF' stands for Frequency of Physical Activity
        physical_activity_col = col
        break

# If a valid column is found, proceed
if physical_activity_col and 'Height' in data.columns and 'Weight' in data.columns:
    # Calculate BMI
    data['BMI'] = data['Weight'] / (data['Height'] ** 2)

    # Create box plot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[physical_activity_col], y=data['BMI'])

    # Add labels
    plt.title('BMI by Physical Activity Levels')
    plt.xlabel('Physical Activity Level')
    plt.ylabel('BMI')

    # Show plot
    plt.show()
else:
    print("Required column for Physical Activity not found. Available columns:", data.columns.tolist())
```

```python
#Group-by Analysis to Summarize and Compare Groups
import pandas as pd

# Define file path
file_path = r"C:\Users\Gulru\Team3_Estimation_of_Obesity_Levels\data\Preprocessed_ObesityDataSet_raw_and_data_sinthetic.csv"

# Load dataset
data = pd.read_csv(file_path)

# Check available columns
print("Available columns:", data.columns.tolist())

# Fix column names dynamically if necessary
if 'Gender' in data.columns and 'Weight' in data.columns:
    # Average weight by gender
    average_weight = data.groupby('Gender')['Weight'].mean()
    print("\nAverage Weight by Gender:\n", average_weight)
else:
    print("\n'Gender' or 'Weight' column not found in dataset.")

# Check if 'FAF' and 'NObeyesdad' exist
if 'FAF' in data.columns and 'NObeyesdad' in data.columns:
    # Convert 'NObeyesdad' to numerical values for analysis
    obesity_mapping = {
        'Insufficient_Weight': 0,
        'Normal_Weight': 1,
        'Overweight_Level_I': 2,
        'Overweight_Level_II': 3,
        'Obesity_Type_I': 4,
        'Obesity_Type_II': 5,
        'Obesity_Type_III': 6
    }
    
    data['Obesity_Level_Num'] = data['NObeyesdad'].map(obesity_mapping)

    # Average obesity level by physical activity level
    average_obesity = data.groupby('FAF')['Obesity_Level_Num'].mean()
    print("\nAverage Obesity Level by Physical Activity Level:\n", average_obesity)
else:
    print("\n'FAF' or 'NObeyesdad' column not found in dataset.")
```

```python
#Group-by Analysis to Summarize and Compare Groups(Graphical Representation)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define file path
file_path = r"C:\Users\Gulru\Team3_Estimation_of_Obesity_Levels\data\Preprocessed_ObesityDataSet_raw_and_data_sinthetic.csv"

# Load dataset
data = pd.read_csv(file_path)

# Convert 'NObeyesdad' to numerical values
obesity_mapping = {
    'Insufficient_Weight': 0,
    'Normal_Weight': 1,
    'Overweight_Level_I': 2,
    'Overweight_Level_II': 3,
    'Obesity_Type_I': 4,
    'Obesity_Type_II': 5,
    'Obesity_Type_III': 6
}
data['Obesity_Level_Num'] = data['NObeyesdad'].map(obesity_mapping)

# 📌 Average weight by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=data.groupby('Gender')['Weight'].mean().index, 
            y=data.groupby('Gender')['Weight'].mean().values, 
            palette='coolwarm')
plt.title('Average Weight by Gender')
plt.ylabel('Weight (kg)')
plt.xlabel('Gender')
plt.show()

# 📌 Average obesity level by physical activity
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data.groupby('FAF')['Obesity_Level_Num'].mean().index, 
                y=data.groupby('FAF')['Obesity_Level_Num'].mean().values, 
                color='blue', alpha=0.6)
plt.title('Obesity Level by Physical Activity Level')
plt.ylabel('Average Obesity Level')
plt.xlabel('Physical Activity (FAF)')
plt.grid(True)
plt.show()
```
