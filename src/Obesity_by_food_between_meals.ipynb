This bar chart shows obesity levels segmented by the frequency of food consumption between meals. 
The x-axis represents different categories of food consumption between meals, while the y-axis represents the percentage of individuals within each obesity level.
The bars are stacked - which is indicating proportion of each obesity level for each category of food consumption.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '../data/Cleaned_ObesityDataSet_raw_and_data_sinthetic.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Ensure the relevant columns exist
if 'CAEC' not in df.columns or 'NObeyesdad' not in df.columns:
    raise ValueError("Required columns 'CAEC' or 'NObeyesdad' not found in the dataset")

caec_obesity_pct = df.groupby(['CAEC', 'NObeyesdad']).size().unstack()
caec_obesity_pct = caec_obesity_pct.div(caec_obesity_pct.sum(axis=1), axis=0) * 100  # Convert to percentages

# Plot the stacked bar chart
caec_obesity_pct.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Obesity Levels by Consumption of Food Between Meals')
plt.xlabel('Consumption of Food Between Meals')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Obesity Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# save a png image
# plt.savefig('obesity_by_food_between_meals.png')

# Display the plot
plt.show()
```
