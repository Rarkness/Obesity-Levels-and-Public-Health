```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('data/Cleaned_ObesityDataSet_raw_and_data_sinthetic.csv')

# Plot histogram for categorical variables with colors representing Obesity Levels
# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
# Create histogram with legend enabled
ax = sns.histplot(data=df, x='Age', hue='NObeyesdad',  legend = True, bins=30, palette="tab10", alpha=0.9, multiple='stack')

plt.title(f'Age Distribution by Obesity Category', fontsize=16, fontweight='bold')
plt.ylabel('Count', fontweight='bold', fontsize=14)
plt.xlabel('Age', fontweight='bold', fontsize=14)
plt.xticks(fontweight='bold', fontsize=12)
plt.yticks(fontweight='bold', fontsize=12)
ax.grid(False)
ax.legend(labels= list(df['NObeyesdad'].unique()), 
          title = 'Obesity Category',
           title_fontsize=18, 
           fontsize='large', 
           loc='upper right', 
           prop={'weight': 'bold', 'size': 12})
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
