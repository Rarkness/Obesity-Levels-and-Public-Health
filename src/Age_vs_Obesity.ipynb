## Age Distribution by Obesity Category
This histogram chart highlights the importance of age as a factor in obesity prevalence and provides valuable insights for healthcare professionals and policymakers to develop targeted interventions.
### Key Observations: 
- **Normal Weight**: The highest proportion is in the early 20s, indicating that younger individuals are more likely to have a normal weight compared to older groups.
- **Overweight & Obesity Levels**: Overweight and obesity levels increase as individuals move from their late teens into their 20s and 30s.Severe obesity types (Obesity Type II and III) are notably present across various ages but are most concentrated in the younger population.
- **Insufficient Weight**: This category is more present in younger individuals, particularly in the early 20s, but diminishes as age increases.
#### Implication
The stacked bars clearly show that while normal weight dominates at younger ages, as individuals grow older, the prevalence of overweight and obesity categories rises.
There is a diverse distribution of obesity levels across different age groups, with a shift from normal weight to obesity in the middle age ranges.




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Load dataset
df = pd.read_csv("G:\DSI\Project\Team3_Estimation_of_Obesity_Levels\data\Cleaned_ObesityDataSet_raw_and_data_sinthetic.csv") 
```

```python
# Plot histogram for categorical variables with colors representing Obesity Levels
# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
# Create histogram with legend enabled
ax = sns.histplot(data=df, x='Age', hue='NObeyesdad',  legend = True, bins=30, palette="tab10", alpha=0.9, multiple='stack')

plt.title(f'Age Distribution by Obesity Category', fontsize=16, fontweight='bold')
plt.ylabel('Count', fontweight='bold', fontsize=14)
plt.xlabel('Age', fontweight='bold', fontsize=14)
plt.xticks(fontweight='bold', fontsize=12)
plt.yticks(fontweight='bold', fontsize=12)
ax.grid(False)
ax.legend(labels= list(df['NObeyesdad'].unique()), 
          title = 'Obesity Category',
           title_fontsize=18, 
           fontsize='large', 
           loc='upper right', 
           prop={'weight': 'bold', 'size': 12})
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
